## Data Preparation

We use 75/15 train-validation split, and monitor F2 score during training since it is the scoring method for the competition.

```python
from sklearn.metrics import fbeta_score
def best_f2(probs,y, start=0., end=1., step=0.01,average='samples'):
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        return max([fbeta_score(y, (probs>th), 2, average=average)
                    for th in np.arange(start,end,step)])

def f2(preds, targs, thresh=0.5): return fbeta(preds, targs, 2, thresh)

def opt_th_single(probs,y, beta=2, start=0., end=1., step=0.01):
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        ths = np.arange(start,end,step)
        idx = np.argmax([fbeta_score(y, (probs>th), beta)
                    for th in ths])
        return ths[idx]
    
def opt_th(probs,y,beta=2, start=0., end=1., step=0.01,average='samples'):
    with warnings.catch_warnings():
        warnings.simplefilter("ignore")
        ths = np.arange(start,end,step)
        idx = np.argmax([fbeta_score(y, (probs>th), beta, average=average)
                    for th in ths])
        return ths[idx]

n = label_df.shape[0]
val_idxs = get_cv_idxs(n,val_pct=0.15,seed=42)
```

```python
augs = [RandomFlip(), 
        RandomRotate(5),
        RandomLighting(0.1,0.1)]
```

```python
arch=resnet34
sz = 224
bs = 16

#test which augmentations are suitable
def get_augs(i,small_batch=10):
    tfms = tfms_from_model(arch, sz, aug_tfms=augs, max_zoom=1.05)
    data = ImageClassifierData.from_csv(DATA_PATH, 'train', label_csv, tfms=tfms,
                    suffix='.jpg', val_idxs=val_idxs, test_name='test')
    x,_ = next(iter(data.aug_dl))
    return data.trn_ds.denorm(x)[i]

ims = np.stack([get_augs(0) for i in range(8)])
plots(ims, rows=2)
```

```python
def get_data(sz):
    tfms = tfms_from_model(arch, sz, aug_tfms=augs, max_zoom=1.05)
    data = ImageClassifierData.from_csv(DATA_PATH, 'train', label_csv, tfms=tfms,
                    suffix='.jpg', val_idxs=val_idxs, test_name='test')
    return data
```

## Training

We first find the optimal learning rate, then train our transfer-learning model on the Planet dataset, freezing all convolution layers. This has already given us quite a decent F2 performance of 0.91799 compared to 0.93317 of the winner. We then start thawing the convolution layers and retrain the model again.

### Find Optimal Learning Rate

```python
arch=resnet34
sz = 224
data = get_data(sz)
opt_fn = partial(optim.Adam, betas=(0.9, 0.999))
learn = ConvLearner.pretrained(arch, data, precompute=False,metrics=[f2,best_f2])
learn.opt_fn = opt_fn
```

```python
learn.lr_find2(start_lr=1e-6, end_lr=1)
learn.sched.plot()
```

```python
lr = 0.25/10
lrm = 2.6 
lrs = np.array([lr/(lrm**3), lr/(lrm**2), lr])
```

### Train with Precomputed Activations

```python
arch=resnet34
sz = 224
data = get_data(sz)
opt_fn = partial(optim.Adam, betas=(0.9, 0.999))
learn = ConvLearner.pretrained(arch, data, precompute=True,metrics=[f2,best_f2])
learn.opt_fn = opt_fn
```

```python
learn.fit(lrs=lr, n_cycle=1, cycle_len=10, use_clr=(10,10))
```

```python
learn.save('precompute')
```

### Train Unfrozen

```python
learn.load('precompute')
learn.precompute=False
learn.unfreeze()
learn.fit(lrs=lrs, n_cycle=1, cycle_len=10, use_clr=(10,10))
```

```python
learn.save('unfrozen')
```

### Quick Results for Agriculture

With precomputed activations, we got best samples F2 score of 0.933028, which is almost as good as the winner of private leaderboard at 0.93317. Since this is a multi-label classification, we cannot simply plot the confusion matrix (there are overlapping labels for one image). Here we take the label `Agriculture` and see how well we perform on identifying all map segments which are agricultural lands.

```python
learn.load('unfrozen')
#get validation performance
probs, y = learn.TTA()
probs = np.mean(probs, 0)
agri_probs = probs[:,0]
agri_y = y[:,0]
th = opt_th_single(agri_probs,agri_y)
convert_pred = np.vectorize(lambda x: 1 if x > th else 0)
agri_preds = convert_pred(agri_probs)
th = 0.5
agri_preds_baseline = convert_pred(agri_probs)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import fbeta_score

print(f'Baseline F2: {fbeta_score(agri_y,agri_preds_baseline,2)}')
print(f'F2: {fbeta_score(agri_y,agri_preds,2)}')
cm = confusion_matrix(agri_y,agri_preds)
plot_confusion_matrix(cm,classes=[0,1])
```
